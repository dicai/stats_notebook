```julia
using Distributions
using Random
using PyPlot
PyPlot.svg(true)
using ColorSchemes
using StatsBase

using Colors
```

```julia
σ2(n) = 1 / (1/4 + n)
μ(n, xbar) = σ2(n) * (1 + xbar)
```

```julia
data = rand(Normal(0, 1), 10000);
```

```julia
n = 1000
xbar(n) = sum(data[1:n])
μ_post = μ(n, xbar(n))
σ_post = σ2(n)
```

```julia
xs = -10:0.01:10

inds = [4, 5, 6, 7,9]
cmap = ColorSchemes.YlGnBu_9
#colors = [join(["#",string(cmap[i], base=16),""]) for i in 1:9];
```

```julia
figure()


PyPlot.tick_params(
    axis="both",          # changes apply to the x-axis
    which="both",      # both major and minor ticks are affected
    bottom=false,      # ticks along the bottom edge are off
    left=false,         # ticks along the top edge are off
    labelbottom=false,
    labelleft=false) # labels along the bottom edge are off

PyPlot.plt.rcParams["font.size"] += 10

axvline(0, c="black", lw=2, ls="dotted")

ys = [pdf(Normal(4,2), x) for x in xs]
plot(xs, ys, c="tab:gray", lw=2)

alphas = [0.35, 0.5, 0.7, 1]
for (i, n) in enumerate([1, 2, 5, 17])
    xbar = sum(data[1:n])
    μ_post = μ(n, xbar)
    σ_post = σ2(n)
    ys = [pdf(Normal(μ_post,σ_post), x) for x in xs]
    plot(xs, ys, c=(0.1+0.1, 0.2, 0.5), alpha = alphas[i], lw=2)
end

PyPlot.xlim(-1.,6)
PyPlot.ylim(0,4.)

legend(["true value", "prior", "posterior 1", "posterior 2", "posterior 3", "posterior 4"])

PyPlot.xlabel("θ")
PyPlot.ylabel("density")

PyPlot.savefig("posterior_consistency.png", tight_layout="true", dpi=200)
```
